```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def f(z, c): 
    return z**2 + c
```

```python
Lx_left = -2.0
Lx_right = 0.75
Ly_down = -1.25
Ly_up = 1.25
Lx = Lx_right - Lx_left
Ly = Ly_up - Ly_down
N = 350
Omega = [[complex((Lx_left + j*Lx/(N-1)), (Ly_up - i*Ly/(N-1))) for j in range(0, N)] for i in range(0, N)]
Omega = np.array(Omega)
Omega
```

```python
K = 10
mand = np.real(np.zeros_like(Omega))

for i in range(0, N):
    for j in range(0, N):
        temp = 0.0+0.0j
        for k in range(0, K):
            temp = f(temp, Omega[i][j])
            if np.abs(temp) > 100:
                mand[i][j] = 1-(K-k)/K
                break;

        
mand
```

```python
temp1 = 0.0+0.0j
for k in range(0, K):
    temp1 = f(temp1, Omega[0][0])
    print(temp1)
```

```python
plt.figure(figsize=(6,6))
plt.xlim(Lx_left, Lx_right)
plt.ylim(Ly_up, Ly_down)
#for i in range(0, N):
#    for j in range(0, N):
for i in range(0, N):
    for j in range(0, N):
        if mand[i][j] == 0.0:
            plt.scatter(np.real(Omega[i][j]), np.imag(Omega[i][j]), color = str(1.0), s=8.)
        else:
            c = mand[i][j]
            plt.scatter(np.real(Omega[i][j]), np.imag(Omega[i][j]), color = str(c), s=2.5)
plt.show()
```
